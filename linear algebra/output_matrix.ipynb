# output matrix

output matrix is a matrix wich each column is a production vector and can be mutliplied by unit cost matrix.if Q is [q1 q2 q3 q4] showing output of each product in

quarters of year,and if unit cost matrix consists of products A , B wich A = (0.45 , 0.25 , 0.15) and B = (0.40 , 0.30 , 0.15) f,the UQ is overall cost of (materials,labor,

overhead) in each quarter

```python
#importing numpy
import numpy as np
```

```python
#creating u

u = np.array( [
    [0.45 , 0.40] ,

    [0.25 , 0.30] ,

    [0.15 , 0.15]
])
```

```python
#Creating Q
q1 = np.array([float(x) for x in input("enter q1: ").split()])

q2 = np.array([float(x) for x in input("enter q2: ").split()])

q3 = np.array([float(x) for x in input("enter q3: ").split()])

q4 = np.array([float(x) for x in input("enter q4: " ).split()])

Q = np.array([q1,q2,q3,q4]).T
```

```python
#calculating uQ
quarters_cost_matrix = u @ Q

print(f"(materials,labor,overhead) cost for ->> first quarter = {quarters_cost_matrix[: , 0]}")

print(f"(materials,labor,overhead) cost for ->> second quarter = {quarters_cost_matrix[: , 1]}")

print(f"(materials,labor,overhead) cost for ->> third quarter = {quarters_cost_matrix[: , 2]}")

print(f"(materials,labor,overhead) cost for ->> fourth quarter = {quarters_cost_matrix[: , 3]}")
```
