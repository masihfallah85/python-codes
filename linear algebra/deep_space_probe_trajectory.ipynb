# deep space probe trajectory

with radio equipment of deep space probe,we get stream of vector data showing how probe position comapres with desired trajectory.if at kth data we assuem X = [x1 x2 .... xk]

we calculate gram matrix as X^t x = GK,its not efficent to compute it each time,isnetad,at time k+1 we can partion X as X = [Xk xk+1] then the gram matrix is Gk+1 = Gk + X(k+1)

^T x(k+1)

```python
#importing numpy

import numpy as np
```

```python
#initial gram matrix

X = np.array([
    [2 , 4] ,

    [5  , 8]
])

X_transpose = X.T

Gram = X_transpose @ X

print(Gram)
```

```python
#for new data 
x_new  = np.array([3, 7])

new_Gram = Gram + ( (x_new.T) @ x_new)

print(new_Gram)
```
