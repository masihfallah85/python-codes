```python
#Imorting numpy and sovle method
import numpy as np

from numpy.linalg import solve
```

```python
#crearing resistor vecotrs(for each current i1,i2,i3 in roder up,middle,buttom loops)and votlage vector

r1 = np.array([11 , -3 , 0])

r2 = np.array([-3 , 6 , -1])

r3 = np.array([0 ,-1 , 3 ])

voltage = np.array([30,5,-25])
```

```python
#Creating coefficent(R)matrix and solving the system(it is a symetric matrix!)
R = np.vstack((r1,r2,r3))

I = solve(R , voltage)
```

```python
#Current values(negative value means its in opposite direction(clockwise here))
print(f"I1 = {I[0]} , I2 = {I[1]} , I3 = {I[2]}")
```
