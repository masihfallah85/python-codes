# matrix partitioning

matrix partitioning can be used for diffrent applications such as parallel processing.we can divide matrix into individual block,additon with a matrix 

with same partition is similar to reguler matrix additon  with blocks,scaler multiplication is also scaler multiplication on blocks.when a matrix has

same column paritioning as rows of another matrix,they are conformabel for matrix mutliplicaiton,with same rules as matrix multiplicatione with blocks.

here we implement an example of matrix multiplication using partitions

```python
#importing numpy
import numpy as np
```

```python
#first matrix
A_1 = np.array([

    [2  , -3 , 1] , 

    [1 , 5 , -2]
])

A_2  = np.array([

    [0 , -4] , 

    [3 , - 1]
])

A_3 = np.array([0 , -4 , -2])

A_4 = np.array([7 , -1])
```

```python
#second matrix
B_1 = np.array([
    [6 , 4] , 

    [-2 , 1] , 

    [-3 , 7]
])

B_2  = np.array([
    [-1 , 3] , 

    [5 , 2] 
])
```

```python
#multiplication

P1 = (A_1 @ B_1) + (A_2 @ B_2)

P2 = (A_3 @ B_1) + (A_4 @ B_2)

AB = np.vstack((P1 , P2))

print(AB)
```
