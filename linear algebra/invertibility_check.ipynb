# invertibility check

in this code we checl if a n*n square matrix is invertible by seeing if it has n pivot positions or not

```python
#importing sympy and numpy
import numpy as np

import sympy as sp
```

```python
#getting matrix
matrix = np.array([float(x) for x in input("enter matrix: ").split()])

n = int(np.sqrt(matrix.size))

matrix = matrix.reshape(n , n)
```

```python
#rref form and getting pivots
matrix = sp.Matrix(matrix)

rred_matrix , pivots = matrix.rref()
```

```python
#checking invertibility 
if len(pivots) < n:

    print("not invertible")

else:

    print("invertible")
```
