# Center of mass vector equation

if we have k amount of points in space,each with  its own mass,called point mass,center of mass is their weigthed average,we can express

it as vector equation  c = 1/ m (m1 * v1 + m2 * v2 + ... mk * vk) wich each vi is vector showing their coordinates and each mi is 

individual point mass,m is sum of all individual masses.

in this problem we have 3 points  , (0 , 1)  , (8 , 1) and ( 2 , 4)  each of have mass of  1 g.now:

1.we want to find center of mass

2.we want to find how to distribute 6 g to each point so center of mass moves to (2 , 2)

```python
#Importing numpy and solve
import numpy as np

from numpy.linalg import solve
```

```python
#vector of each individual point
point_a = np.array([0 , 1] , dtype=np.float64)

point_b = np.array([8 , 1] , dtype=np.float64)

point_c = np.array([2 , 4] , dtype=np.float64)

print(f"point_a : {point_a} , point_b: {point_b} , point_c: {point_c}")
```

```python
#Calculate center of mass
center_of_mass = (1 / 3) * (1 * point_a + 1 * point_b + 1 * point_c)

print(f"center of mass of {point_a} with mass of 1 g, {point_b} with mass of 1 g , {point_c} with mass of 1 g : {center_of_mass}")
```

```python
#Finding how to distribute 6 g
center_of_mass = np.array([2 , 2] , dtype=np.float64)

center_of_mass *= 9

#Since sum of masses equal 9,we can right one point based on others,so we remove first point from equation
center_of_mass -= (9 * point_a)

point_b -= point_a

point_c -= point_a

#solving system 
coefficent_matrix = np.vstack((point_b , point_c))

coefficent_matrix = coefficent_matrix.T

distribution = solve(coefficent_matrix , center_of_mass)

distribution_of_b = distribution[0] - 1

distribution_of_c = distribution[1] - 1

distribution_of_a = (9 - (distribution_of_b + distribution_of_c + 2 )) - 1

print(f"add {distribution_of_a} g to point a , add {distribution_of_b} g to point b , add {distribution_of_c} to point c")
```
