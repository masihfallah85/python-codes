# useful linear transformations

here we have an example of useful transformations expansion,contracion,projection,rotation,shear and reflection,and their onto,one to one properties with R^2 as domain.

here standard matrices are standard matrices of transformations and t_e1 or t_e2 is what transformation do to x,y axis,required for cosntruction of standard matrix

it is one to one if it doesnt have free variables and onto if it's columns spans R^m(here R^2)

```python
#importing numpy

import numpy as np

from numpy.linalg import matrix_rank
```

```python
#getting point

v = np.array([float(x) for x in input("enter point: ").split()])

print(v)
```

```python
#contraction and dialation

k = float(input("enter scale of contraction or expansion: "))

T1_e1 = np.array([k , 0])

T2_e2 = np.array([0,1])

standard_matrix_1 = np.array([T1_e1 , T2_e2]).T

T2_e1 = np.array([1 , 0])

T2_e2 = np.array([0 , k])

standard_matrix_2 = np.array([T2_e1 , T2_e2]).T

T3_e1 = np.array([k , 0])

T3_e2 = np.array([0 , k])

standard_matrix_3 = np.array([T3_e1 , T3_e2]).T

print(f"vector under horizontal contraction/expansion: {standard_matrix_1 @ v}")

print(f"vector under vertizal contraction/expansion: {standard_matrix_2 @ v}")

print(f"vector contraction/expansion {standard_matrix_3 @ v}")

```

```python
#projection 

axis  = input("enter wich axis to project into: ")


if axis.lower() == "x":

    t_e1 = np.array([1 , 0])

    t_e2 = np.array([0 , 0])

elif axis.lower() == "y":

    t_e1 = np.array([0 , 0])

    t_e2 = np.array([0 , 1])

standard_matrix_3= np.array([t_e1 , t_e2]).T

print(f" projection of vector to {axis} axis = {standard_matrix_3 @ v}")
```

```python
#rotation

phi = float(input("enter degree of rotation: "))

degree = phi

phi = np.deg2rad(phi)

counter_clockwise = phi 

clockwise = -1 * phi

t1_e1 = np.array([np.cos(counter_clockwise) , np.sin(counter_clockwise)])

t1_e2 = np.array([-1 * np.sin(counter_clockwise) , np.cos(counter_clockwise)])

standard_matrix_4 = np.array([t1_e1 , t1_e2])

t2_e1 = np.array([np.cos(clockwise) , np.sin(clockwise)])

t2_e2 = np.array([-1 * np.sin(clockwise) , np.cos(clockwise)])

standard_matrix_5 = np.array([t2_e1 , t2_e2])

print(f"counter clockwise rotation with {degree} degree : {standard_matrix_4 @ v}")

print(f" clockwise rotation with {degree} degree : {standard_matrix_5 @ v}")
```

```python
#shear

k = float(input("enter shear scale: "))

t1_e1 = np.array([1 , 0])

t1_e2 = np.array([k , 1])

standard_matrix_6 = np.array([t1_e1 , t1_e2]).T


t2_e1 = np.array([1 , k])

t2_e2 = np.array([0 , 1])

standard_matrix_7 = np.array([t2_e1 , t2_e2]).T

print(f"vector under horizontal shear: {standard_matrix_6 @ v}")

print(f"vector under vertical shear: {standard_matrix_7 @ v}")
```

```python
#reflection
t1_e1 = np.array([-1, 0])

t1_e2 = np.array([0, 1])

standard_matrix_8 = np.array([t1_e1 , t1_e2]).T

t2_e1 = np.array([1, 0])

t2_e2 = np.array([0, -1])

standard_matrix_9 = np.array([t2_e1 , t2_e2]).T

t3_e1 = np.array([0, 1])

t3_e2 = np.array([1, 0])

standard_matrix_10 = np.array([t3_e1 , t3_e2]).T

t4_e1 = np.array([-1, 0])

t4_e2 = np.array([0, -1])

standard_matrix_11 = np.array([t4_e1 , t4_e2]).T

t5_e1 = np.array([0, -1])

t5_e2 = np.array([-1, 0])

standard_matrix_12= np.array([t5_e1 , t5_e2]).T

print(f"vector reflected through y axis = {standard_matrix_8 @ v}")

print(f"vector reflected through x axis = {standard_matrix_9 @ v}")

print(f"vector reflected through y=x = {standard_matrix_10 @ v}")

print(f"vector reflected through origin = {standard_matrix_11 @ v}")

print(f"vector reflected through y=-x = {standard_matrix_12 @ v}")
```

```python
#onto,one to one properties
matrices = [standard_matrix_1 , standard_matrix_2 , standard_matrix_3 , standard_matrix_4 , standard_matrix_5 , standard_matrix_6 , standard_matrix_7  , standard_matrix_8 , standard_matrix_9 , standard_matrix_10 , standard_matrix_11 , standard_matrix_12]


for i in range(1 , 13):

    rank = matrix_rank(matrices[i - 1])

    if rank < len(matrices[i - 1]):

        print(f"standard_matrix{i} is not onto")

    else:

        print(f"standard_matrix{i} is onto")

    if rank < len(matrices[i - 1][0]):

        print(f"standard_matrix{i} is not one to one")

    else:

        print(f"standard_matrix{i} is one to one")
```
