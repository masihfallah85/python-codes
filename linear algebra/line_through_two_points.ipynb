# Line through two points

by giving coordinates of two points,we can give parametric descirption of lien throug them,simply add one of them to multiplies of

their difference,as a it is a line through them parallel to their difference

```python
#Importing numpy

import numpy as np
```

```python
#Getting vectors

dimensionality = int(input("enter dimensionality: "))


p = []

q = []

for i in range(dimensionality):

    p.append(float(input("enter entries of p: ")))


for i in range(dimensionality):

    q.append(float(input("enter entries of q: ")))

p = np.array(p)

q = np.array(q)
```

```python
#parametric description of lien through them
v   = p - q

print(f"line through {q} and {p} : x1 * {v} + {p}")
```
