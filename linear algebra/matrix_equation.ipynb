# Matrix equation

any vector equation can be translated to matrix equation,if we have a1,a2..an vectos and x1,x2..xn weights,then matrix [a1 a2 ... an] 

mutliplied  by a vector x^t = (x1,x2....xn) is same as x1 * a2 + x2 * a2 +.....xn * an,wich here the matrix acts as coefficent matrix of 

system  of linear equations

```python
#Importing numpy and solve function
import numpy as np

from numpy.linalg import solve
```

```python
#Creating coefficent matrix

coefficent_matrix  = []

m = int(input("enter number of rows: "))

n = int(input("enter number of columns: "))

for i in range(n):

    column = []

    for j in range(m):

        column.append((float(input("enter entrie: "))))

    coefficent_matrix.append(column)

coefficent_matrix = np.array(coefficent_matrix)

coefficent_matrix = coefficent_matrix.T

print(coefficent_matrix)
```

```python
#Getting weigths vector

weigths = []

for i in range(n):

    weigths.append(float(input("enter weigth: ")))


weigths = np.array(weigths)

print(weigths)
```

```python
#Getting augmented column

augmented_column = []

for i in range(m):

    augmented_column.append(float(input("enter augmented entrie: ")))

augmented_column = np.array(augmented_column)

print(augmented_column)
```

```python
#Multiplying matrix with weigths vector

result = []

for i in range(m):

    addition = 0

    for j in range(n):

        addition += (coefficent_matrix[i , j] *  weigths[j])

    result.append(addition)

print(result)

#Using numpy's own method

result_2 = coefficent_matrix @ weigths

print(result_2)
```

```python
#Solving matrix equation 

result = solve(coefficent_matrix , augmented_column)

print(result)
```
