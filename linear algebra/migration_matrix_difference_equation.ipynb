# migeration matrix difference equation

we implement a simple 2 * 2 migeration matrix wich indicates what portion of population go from where to where,here we use inside and outside of city for each year.

since each year relies on previous year this creates a linear difference equation.

```python
#Importing numpy
import numpy as np

from numpy.linalg import matrix_power

from math import ceil
```

```python
#getting iniall count of people in city and outside city
initial = np.array([int(x) for x in input("enter count of people as : inside city , outside city with space: ").split()])
```

```python
#creating migration matrix
inside = np.array([float(x) for  x in input("enter wich portion of population remain inside city and wich portion goes outsie with space between them: ").split()])

outside = np.array([float(x) for  x in input("enter wich portion of population remain outside city and wich portion come inside  with space between them: ").split()])

migration_matrix = np.array([inside , outside]).T
```

```python
#Population after n years
n = int(input("enter how many years from now do you want to know population of city and outside of it: "))

migration_matrix_power = matrix_power(migration_matrix , n)

population = migration_matrix_power @ initial

print(f"population of city ={ceil(population[0])} , population of outside = {ceil(population[1])}")
```
