# Linear_transformation application

a transformation is linear if T(0) = 0 (can be deduced from T(uX) = u T(X)) and that T(c1v1+c2v2+..cnvn) = c1T(v1) + c2T(v2) ... + cn T(vn)(can be deduced from T(u+v) = T(u) + T

(v) ) wich is superposition principle,meaning  output of input to system as linear combination of signals by a magnitute is sum of each output of individual signals multiplied by

their magnitued.

matrix equation Ax is also a linear transformation from domain r^n to codomain R^m with a range,forming m * n matrix.a dynamical view of matrix * vector.

here we have an example of linear transformation view application:

if a company produces two products a and b,and cost per dollar in format of (materials,labor,overhead) for a is (0.45 , 0.25 , 0.15) and for b (.40,0.30,0.15) ,then calculate

overall expenses with x dollars of a and y dollars of b.

```python
#Importing numpy
import numpy as np
```

```python
#Getting x and y.

x = float(input("enter how much dollars of product a: "))

y = float(input("enter how much dollars of product b: "))

production_vector = np.array([x,y])
```

```python
#unit cost matrix

a = np.array([0.45,0.25,0.15])

b = np.array([0.40,0.30,0.15])

unit_cost_matrix = np.array([a , b]).T
```

```python
#overall expenses
expenses_vector = unit_cost_matrix @ production_vector

print(f"materials cost = {expenses_vector[0]} , labor cost = {expenses_vector[1]} , overhead = {expenses_vector[2]}")
```
