# givens rotation

givens rotation is a rotation used to create zero entries in vector(usually column of a matrix),wich it's determinat always equals 1.here we have a

givens rotation in R^3

```python
#importing numpy

import numpy as np

from numpy import sqrt
```

```python
#standard matrix of transformation
givens_rotation = np.array([
    [sqrt(5) / 5 , 0 , (2 * sqrt(5)) / 5],
    [0,1,0] , 
    [(-2 * sqrt(5)) / 5 , 0 , sqrt(5) / 5 ]
])
```

```python
#getting vector
vector = np.array([float(x) for x in input("enter vector: ").split()])
```

```python
#result
result  = givens_rotation @ vector

print(result)
```
