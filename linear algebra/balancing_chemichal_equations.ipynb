# balancing chemichal equations

we can turn each compund as a vector based on elements,then solve homogeneous system of linear equations and find integer coefficent for chemichal equation.

here we have this equation as an example:

c3h8 + o -> co2 + h2o

```python
#Importing numpy

import numpy as np

from math import ceil
```

```python
#compounds
c3h8 = np.array([3,8,0],dtype = np.float64)

o = np.array([0,0,1] , dtype=np.float64)

co2 = np.array([1,0,2] , dtype =np.float64)

h2o = np.array([0,2,1] , dtype=np.float64)

print(f"c3h8 vector form = {c3h8}")

print(f"o vector form = {o}")

print(f"co2 vector form = {co2}")

print(f"h2o vector form = {h2o}")
```

```python
#forming augmented matrix

co2 *= -1

h2o *= -1

augmented_matrix_transpose = np.vstack((c3h8 , o , co2 , h2o ))

augmented_matrix = augmented_matrix_transpose.T

augmented_matrix = np.hstack((augmented_matrix , np.zeros((3 , 1))))

print("augmented matrix: " , augmented_matrix)
```

```python
#solving system
augmented_matrix[1] += ( ( -8 / 3) * augmented_matrix[0] )

r = augmented_matrix[2].copy()

augmented_matrix[2] = augmented_matrix[1].copy()

augmented_matrix[1] = r

augmented_matrix[2] *= (( -1 / 2.66666667  ))

augmented_matrix[0] *= (( 1 / 3))

augmented_matrix[2] *= -1

r = augmented_matrix[2].copy()

r *= 2

r1 = augmented_matrix[1].copy()

r2  = np.array([x + y for x , y in zip(r , r1)])

r2[2] = 0

r2[3] = -1.25

augmented_matrix[1] = r2

r = augmented_matrix[2].copy()

r *= (0.33333333)

r1 = augmented_matrix[0].copy()

r2  = np.array([x + y for x , y in zip(r , r1)])

r2[2] = 0

augmented_matrix[0] = r2

print(augmented_matrix)
```

```python
#solution
solution_vector = np.array([-1 * augmented_matrix[0 , 3] , -1 * augmented_matrix[1 , 3] , -1 * augmented_matrix[2,3] , 1])

solution_vector *= 4

print(f"{ceil(solution_vector[0])}c3h8 + {solution_vector[1]}o -> {ceil(solution_vector[2])}co2 + {solution_vector[3]}h2o")
```
