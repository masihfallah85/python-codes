# invertible matrix application (flexibility matrix)

we can use invertible matrixes for solving linear system of equations,one example here is flexibility matrix,if a horizental elastic beam is subjected to forces at n points,

then by hooks law deflectionsare D*f wich f is vector fo forces at eahc poitn and D is fleixbiltuy matrix,it's inverse is called stiffnes matrix,if we have deflections,

uisng it we can find individual forces at each point

```python
#Importing numpy
import numpy as np
```

```python
#getting deflections
deflections  = np.array([float(x) for x in input("enter deflections of each point: ").split()])
```

```python
#getting flexibility matrix

flexibility_matrix = np.array([float(x) for x in input("enter flexibility matrix: ").split()])

n = int(np.sqrt(flexibility_matrix.size))

flexibility_matrix = flexibility_matrix.reshape(n,n)
```

```python
#stiffness matrix
stiffness_matrix = np.linalg.inv(flexibility_matrix)
```

```python
#forces vector 
forces = stiffness_matrix @ deflections

for i in range(len(forces)):

    print(f"force at point {i + 1} = {forces[i]}")
```
