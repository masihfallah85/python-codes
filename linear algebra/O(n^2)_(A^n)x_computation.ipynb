# o(n^2) (A^n)x computation

insetad of calculating A^n then multipling x with it,o(n^3) algorithm,we decompose A^n to a series fo n,with property of associativity of matrix multiplication,and each time


based on definiton of matrix multiplication  we create new vector as sum of columns of A using x entries as weigths,an o(n^2) algorithm

```python
#importing numpy
import numpy as np
```

```python
#getting x
x = np.array([float(i) for i in input("enter x: ").split()])
```

```python
#getting A
A = np.array([float(i) for i in input("enter A: ").split() ] ).reshape(3,3)
```

```python
#getting n

n = int(input("enter n " ))
```

```python
#calculating A^n x
for i in range(n):

    z = np.zeros(3)

    for j in range(len(x)):

        z += (x[j] * A[:,j])

    x = z

print(x)
```
