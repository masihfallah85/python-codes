# shur complement

in this script we implement how to calculate shur compelemnt with an example statistical application 

assume we have a data matrix X and then we add new predictor x0 to form W = [X x0] partitioning

```python
#importing numpy


import numpy as np
```

```python
#constructing w

n = 3

m = 10

np.random.seed(0)

X  = np.random.randn(m , n)

x0 = np.random.randn(m , 1)

W = np.hstack([X  , x0])

print(W)
```

```python
#Computing W^t w with paritoning [X x0]

A11 = X.T @ X

A12 = X.T @ x0

A21 = x0.T @ X

A22 = x0.T @ x0
```

```python
#shur complement

A11_inverse = np.linalg.inv(A11)

S = A22 - A21 @ A11_inverse @ A12

print(S)
```

```python
#veirfy using projection matrix M

M = np.eye(m) - (X @ A11_inverse @ X.T)

s_check = x0.T @ M @ x0

print(s_check)
```

```python
#relation to W^t w ( 1 / s is porportional to variance )

WtW_inverse = np.linalg.inv(W.T @ W)

print(WtW_inverse[-1 , -1])

print(f"using S = {1 / S}")
```
