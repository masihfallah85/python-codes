#  nutritious weight loss diet construcion

in this script,we use linear algebra to construct a nutritious Weight loss diet,here we need this amount  of (protein , carbohydrate,fat) in grams -> (33,45,3)

we have 3 ingredients,nonfat milk,soy flour and whey,each one having these (protein , carbohydrate,fat) in per 100 grams of them:

nonfat milk -> (36,52,0)

soy flaur ->(51,31,7)

whey -> (13,74,1.1)

we know that if we take each ingredients material per 100 g as a vector,x scaler multiplied by vector gives us amount of material in x unit(100 g is our unit) of ingredient.

for general case,we need to be sure that solution to this question doesn't give us negative amount of ingridients.

```python
#Importing numpy
import numpy as np

from numpy.linalg import solve
```

```python
#constructing unit materials matrix for ingridents

nonfat_milk = np.array([36,52,0] , dtype = np.float64)

soy_flaur = np.array([51,31,7] , dtype= np.float64)

whey = np.array([13,74,1.1])

unit_material_matrix = np.array([nonfat_milk, soy_flaur, whey]).T

print(unit_material_matrix)
```

```python
#Creating augmented column

amount_of_materials = np.array([33,45,3] , dtype=np.float64)

print(amount_of_materials)
```

```python
#Solution 
amount_of_ingridients = solve(unit_material_matrix , amount_of_materials)

print(f"amount of nonfatmilk = {amount_of_ingridients[0]} , amount of soy flaur = {amount_of_ingridients[1]} , amount of whey = {amount_of_ingridients[2]} (in unit of 100 g) ")
```
