# mining operation vector equation

In this script we use a mining operation example for application of vector equation

In two mines,mine A in one day produces 20 metric tons of copper and 550 kilograms of silver,mine b in one day   produces ore that 

contains 30 metric tons of copper and 500 kilograms of silver.in how many days,both mines  produce 150 tons of copper and 2825 kilograms

of silver?

```python
#Importing numpy and solve
import numpy as np

from numpy.linalg import solve
```

```python
#Intepert each mines produced material in a day as a vecotor,in form of (copper , silver)
mine_A = np.array([20 , 550])

mine_B = np.array([30 , 500])

goal = np.array([150 , 2825])
```

```python
#Forming coefficent matrix
coefficent_matrix = np.vstack((mine_A , mine_B))

coefficent_matrix = coefficent_matrix.T
```

```python
#Solving the problem
days = solve(coefficent_matrix , goal)

print(f"{days[0]} days at mine A and  {days[1]} days at mine B")
```
