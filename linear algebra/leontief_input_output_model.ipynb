# Leontief input_output model 

if economics is divided into multiple sectors,if we have total output in a year and how it is divided to other sectors,we can find an eqalibrium outputs(prrices) for sectors

such that for all sectors expanse/income balances out.

as an example suppose we have 3 sectors,coal,electricity and steel,if electricty pruchase 60%  coal,20% steel and retain rest,and if coal pruchase 40% to electricty and 

60%  steel,and if steel pruchase 40%  coal and 50%  electricity and retain rest,find this equlibrium prices(we transfrom it to homogeneous equations)

```python
#Importing numpy
import numpy as np
```

```python
#Creating values of equations(pe electricity,pc coal,and ps steel)
pc_1 = 1

pe_1 = -0.4

ps_1 = -0.6

pc_2 = -0.6

pe_2 = 1 - 0.1

ps_2 =  -0.2

pc_3 = -0.4

pe_3 = -0.5

ps_3 = 1 - 0.2
```

```python
#Creating augmented matrix
augmented_matrix =  np.array([

    [pc_1 , pe_1 , ps_1 , 0] ,

    [pc_2 , pe_2 , ps_2 , 0] , 

    [pc_3 , pe_3 , ps_3 , 0]
])

print(augmented_matrix)
```

```python
#Solving it
augmented_matrix[1] += 0.6 * augmented_matrix[0]

augmented_matrix[2] += 0.4 * augmented_matrix[0]

augmented_matrix[2] += augmented_matrix[1]

augmented_matrix[1] *= ( 1 / 0.66)

augmented_matrix[0] += 0.4 * augmented_matrix[1]

print(augmented_matrix)
```

```python
#Creating span of solutions
vector = np.array([(-1 * augmented_matrix[0 , 2]) , (-1 * augmented_matrix[1 , 2]) , 1])

print(f"Solutions = span of {vector}")
```

```python
#Check 
ps = float(input("Enter steel price: "))

vector *= ps

ps = vector[2]

pe = vector[1]

pc = vector[0]


print(f"pc = {pc} , pe= {pe} , ps = {ps}")
```
