# Vector equation

we can describe system of linear equations as vector equation,as columns of coefficent matrix are vectors,wich variabels serve as weigths.

if x1 , x2 ..... xn are variabels and a1 , a2 ,.....an are columns of coefficent matrix and b is augmented column,the vector equation

x1 * a1 + x2 * a2 + x3 * a3  ..... = b

wich checks wethever b is in span{a1 , a2 ,...an} or not.span of set of vectors is their linear combination

note :if  system of linear equations is overdetermined or underdetermined,solve method gives error,so this code works for a 

square coefficent matrix.

```python
#Importing numpy and solve function

import numpy as np 

from numpy.linalg import solve
```

```python
def span(set_of_vectors : list , augmented_column : np.array):

    """This functions checks if a vector is in span of set of vectors or not,if it is it prints the linear combination"""


    coefficent_matrix = []

    for i in set_of_vectors:

        coefficent_matrix.append(i)

    coefficent_matrix = np.array(coefficent_matrix)

    coefficent_matrix = coefficent_matrix.T

    try:

      print("coefficent matrix: " , coefficent_matrix)

      print()

      print("augmented_column: " , augmented_column)

      print()
    
      weights = solve(coefficent_matrix , augmented_column)

      print(f"wieghts = {weights}")
        
      print(f"{weights[0]} *  {set_of_vectors[0]}  " , end = "")

      for i in range(1 , len(set_of_vectors)):

          print(f" + {weights[i]} * {set_of_vectors[i]}" , end = "")

      print( f"= {augmented_column}")

      print()
    except Exception as e:

        print(f"{augmented_column} is not in span {set_of_vectors}")
```

```python
#Getting input vectors

set_of_input_vectors = []

amount = int(input("amount of vectors in set: "))

dimensionality = int(input("dimensionality of each vector: "))

for i in range(amount):

    vector = []

    for j in range(dimensionality):

        vector.append(float(input("Enter entry: ")))

    set_of_input_vectors.append(vector)

print(set_of_input_vectors)
```

```python
#Getting augmented vector

augmented_vector = []

for i in range(dimensionality):

    augmented_vector.append(float(input("Enter augmented vector entrie: ")))

augmented_vector = np.array(augmented_vector)

print(augmented_vector)
```

```python
#Solving the equation
span(set_of_input_vectors , augmented_vector)
```
