# Linear_independence_check

if we have set of n vectors,we can form this linear relation:

c1v1+c2v2 ....cnvn=0 

if it has only travial solution(all weigths are zero),then they are lineary dependent and above equation is their linear dependence relationship,if not then they are

lineary indepedent,we can say this as whenever one vector is inside span of others or not geometricly.

by using them as columns of matrix,we can say same things abot that matrix,and through it's rank,check if they are lineary independent(rank = n) or not

```python
#Importing numpy
import numpy as np

from numpy.linalg import matrix_rank
```

```python
#Getting vectors

lst = []

n = int(input("enter number of vectors: "))

for i in range(n):

    v =  input().split()

    v = [float(x) for x in v]

    v = np.array(v)

    lst.append(v)

lst = np.array(lst)
```

```python
#Linear independence check


if matrix_rank(lst) == n:

    print("vectors are independent")

else:

    print("vectors are dependent")
```
